### Confusion Matrixes

Lastly, we want to see how well the model predicts. Since we have imbalanced target we will use the confusion matrix to give us insight on how well the model performs

```python
# Confusion Matrix - Dataset with outliers
for name, classifier_outlier in classifiers_outlier:
    y_outlier_pred = classifier_outlier.predict(X_outlier_test)
    cm = confusion_matrix(y_outlier_test, y_outlier_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Other', 'Low', 'Middle', 'High', 'Luxury'])
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix for {name}")
    plt.show()
```

```python
# Confusion Matrix - Dataset without outliers (IQR Treatment)
for name, classifier_IQR in classifiers_IQR:
    y_IQR_pred = classifier_IQR.predict(X_IQR_test)
    cm = confusion_matrix(y_IQR_test, y_IQR_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Other', 'Low', 'Middle', 'High', 'Luxury'])
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix for {name}")
    plt.show()
```

```python
# Confusion Matrix - Dataset without outliers (Winsorisation Treatment)
for name, classifier_winsor in classifiers_winsor:
    y_winsor_pred = classifier_winsor.predict(X_winsor_test)
    cm = confusion_matrix(y_winsor_test, y_winsor_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Other', 'Low', 'Middle', 'High', 'Luxury'])
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix for {name}")
    plt.show()
```

The confusion matrixes within these three different datasets produce rather interesting insight.
The model performs well on the Middle class but struggles with the other four. Class imbalance is likely an issue as the Middle class has many more instances than the other classes, which affects the overall performance.

## Saving the Model

Finally, we are saving the model. We will be saving three into three different files since we applying several models into different datasets (df_nyc_backup, df_nyc_IQR, df_nyc_winsor)

```python
trained_models_outlier = {}
for name, classifier_outlier in classifiers_outlier:
    print(f"Training {name}...")
    classifier_outlier.fit(X_outlier_train, y_outlier_train)
    trained_models_outlier[name] = classifier_outlier
```

```python
trained_models_IQR = {}
for name, classifier_IQR in classifiers_IQR:
    print(f"Training {name}...")
    classifier_IQR.fit(X_IQR_train, y_IQR_train)
    trained_models_IQR[name] = classifier_IQR
```

```python
trained_models_winsor = {}
for name, classifier_winsor in classifiers_winsor:
    print(f"Training {name}...")
    classifier_winsor.fit(X_winsor_train, y_winsor_train)
    trained_models_winsor[name] = classifier_winsor
```

```python
with open('trained_models_outlier.pkl', 'wb') as file:
    pickle.dump(trained_models_outlier, file)

with open('trained_models_IQR.pkl', 'wb') as file:
    pickle.dump(trained_models_IQR, file)

with open('trained_models_winsor.pkl', 'wb') as file:
    pickle.dump(trained_models_winsor, file)
```

We have successfuly saved the model into three different file. <br> Thus making it recycleable for future inplementation on different test dataset.
